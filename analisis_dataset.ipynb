**Pregunta 3:** Dado el dataset "estudiantes.csv":

* Agregar la columna promedio que contenga el promedio de las 3 notas.
* Utiliza el método query() para filtrar el DataFrame y seleccionar solo las filas donde el nombre del estudiante comienza con una letra específica (por ejemplo, 'A') y donde haya obtenido una calificación superior al promedio en al menos dos asignaturas.
* Almacena el resultado filtrado en un nuevo DataFrame llamado calificaciones_filtradas.
* Muestra el DataFrame calificaciones_filtradas.

```python
# Leer el CSV
df = pd.read_csv(
    "estudiantes.csv",
    sep="-",
    header=0,
    names=["id", "Nombre", "Asignatura1", "Asignatura2", "Asignatura3"]
)

# Convertir las notas a numéricas
cols_notas = ["Asignatura1", "Asignatura2", "Asignatura3"]
df[cols_notas] = df[cols_notas].astype(int)

# Calcular el promedio de las 3 asignaturas
df["promedio"] = df[cols_notas].mean(axis=1)

# Filtrar usando SOLO query()
calificaciones_filtradas = df.query(
    'Nombre.str.startswith("A") and '
    '((Asignatura1 > promedio)*1 + (Asignatura2 > promedio)*1 + (Asignatura3 > promedio)*1) >= 2',
    engine="python"
)

# Mostrar el DataFrame resultante
calificaciones_filtradas
```

**Pregunta 5:** Utilizando el dataset anterior añadir la columna "Categoria" que se cree a partir de la columna "Promedio" con las siguientes categorias:

* 0-70 >> 'Necesita mejorar'

Acontinuación:

* Imprimir cuantos alumnos tienen una nota Aceptable

```python
# Crear la columna Categoria a partir de Promedio
bins = [0, 70, 80, 90, 100]
labels = ["Necesita mejorar", "Aceptable", "Bueno", "Excelente"]

df["Categoria"] = pd.cut(
    df["promedio"],
    bins=bins,
    labels=labels,
    include_lowest=True,
    right=False
)

# Imprimir cuántos alumnos tienen nota Aceptable
print("Alumnos con nota Aceptable:",
      (df["Categoria"] == "Aceptable").sum())

# Imprimir cuántos alumnos tienen nota Excelente
print("Alumnos con nota Excelente:",
      (df["Categoria"] == "Excelente").sum())
```

**Pregunta 7:** Dado los datasets de ventas de enero y febrero:

* Crea un dataframe por cada dataset.
* Añadir la columna mes a cada dataframe que contenga el valor del mes.
* Concatenar ambos datasets a travez del eje 0

Utilizando GroupBy

* Obtener la cantidad total vendida del producto A y B.
* Obtener los ingresos totales en el mes de febrero.

```python
datos_enero = {
    'Producto': ['A', 'B', 'C'],
    'Cantidad': [100, 150, 200],
    'Ingresos': [1000, 2000, 3000]
}

datos_febrero = {
    'Producto': ['A', 'B', 'D'],
    'Cantidad': [120, 180, 220],
    'Ingresos': [1200, 2200, 3300]
}
```

```python
# Datasets
datos_enero = {
    "Producto": ["A", "B", "C"],
    "Cantidad": [100, 150, 200],
    "Ingresos": [1000, 2000, 3000]
}

datos_febrero = {
    "Producto": ["A", "B", "D"],
    "Cantidad": [120, 180, 220],
    "Ingresos": [1200, 2200, 3300]
}

# 1) Crear un DataFrame por cada dataset
df_enero = pd.DataFrame(datos_enero)
df_febrero = pd.DataFrame(datos_febrero)

# 2) Añadir la columna mes
df_enero["mes"] = "enero"
df_febrero["mes"] = "febrero"

# 3) Concatenar ambos DataFrames por el eje 0
df_ventas = pd.concat([df_enero, df_febrero], axis=0, ignore_index=True)

# 4) GroupBy: cantidad total vendida de los productos A y B
cantidad_AB = (
    df_ventas[df_ventas["Producto"].isin(["A", "B"])]
    .groupby("Producto", as_index=False)["Cantidad"]
    .sum()
)

# 5) GroupBy: ingresos totales en el mes de febrero
ingresos_febrero = (
    df_ventas[df_ventas["mes"] == "febrero"]
    .groupby("mes", as_index=False)["Ingresos"]
    .sum()
)

# Mostrar resultados
print("DataFrame concatenado de ventas:")
print(df_ventas)

print("\nCantidad total vendida por producto (A y B):")
print(cantidad_AB)

print("\nIngresos totales en febrero:")
print(ingresos_febrero)
```

**Pregunta 10:** Dado el dataset titanic:

* Categorizar la columna edad en Infantes,Niños,Adolescentes,Jóvenes,Adultos, Adulto Mayor y SinDato.
* Crear un nuevo dataframe que contenga la edad y las categorías.
* Apartir de la dataframe anterior obtener que porcentaje de pasajeros fueron Infantes,Niños,Adolescentes,Jóvenes,Adultos, Adulto Mayor y SinDato.
* Crear un nuevo dataframe que contenga las categorias,conteo,porcentaje, como se muestra a continuación.

```python
# Leer titanic.csv 
titanic_df = pd.read_csv("titanic.csv")
titanic_df.columns = titanic_df.columns.str.strip()  # por si hay espacios

# Categorizar la edad (Age)
bins = [0, 2, 12, 18, 30, 60, 120]
labels = ["Infantes", "Niños", "Adolescentes", "Jóvenes", "Adultos", "Adulto Mayor"]

titanic_df["Categoria"] = pd.cut(titanic_df["Age"], bins=bins, labels=labels, include_lowest=True)
titanic_df["Categoria"] = titanic_df["Categoria"].cat.add_categories("SinDato").fillna("SinDato")

# DataFrame solicitado con Age y Categoria
df_edades = titanic_df[["Age", "Categoria"]]

df_edades
```

```python
# Conteo por categoría
conteo = df_edades["Categoria"].value_counts()
porcentaje = (conteo / conteo.sum() * 100).round(2)

df_resultado = pd.DataFrame({
    "Categoria": conteo.index,
    "Conteo": conteo.values,
    "Porcentaje (%)": porcentaje.values
})

# ORDEN de las categorías
orden = [
    "Infantes",
    "Niños",
    "Adolescentes",
    "Jóvenes",
    "Adultos",
    "Adulto Mayor",
    "SinDato"
]

df_resultado = df_resultado.set_index("Categoria").loc[orden].reset_index()

df_resultado
```

**Pregunta 12:** Dado el dataset titanic:

* Se desea saber que porcentaje de sobrevivientes fueron mujeres(en general).
* Se desea saber que porcentaje de sobrevivientes fueron infantes-niños(0-12 años).


```python
# Leer el dataset titanic
df = pd.read_csv("titanic.csv")
df.columns = df.columns.str.strip()  # por si hay espacios en nombres

# Filtrar solo sobrevivientes
surv = df[df["Survived"] == 1]
total_surv = len(surv)

# % de sobrevivientes que fueron mujeres 
conteo_sex = surv.groupby("Sex").size()
porc_mujeres = (conteo_sex.get("female", 0) / total_surv * 100).round(2)

# % de sobrevivientes que fueron infantes-niños (0-12)
ninos_surv = surv[(surv["Age"] >= 0) & (surv["Age"] <= 12)]
porc_ninos = round(len(ninos_surv) / total_surv * 100, 2)

print(f"Porcentaje de sobrevivientes mujeres: {porc_mujeres}%")
print(f"Porcentaje de sobrevivientes infantes-niños (0-12): {porc_ninos}%")
```

**Pregunta 13:** Dado el dataset titanic:

* Se desea saber la cantidad de personas sobrevivientes y no sobrevivientes en primera clase.
* A partir del resultado anterior se desea saber la cantidad de hombres y mujeres en ambos casos.

```python
# Leer dataset
df = pd.read_csv("titanic.csv")
df.columns = df.columns.str.strip()

# Filtrar solo primera clase
primera = df[df["Pclass"] == 1]

# Cantidad de sobrevivientes y no sobrevivientes en primera clase
conteo_surv = primera["Survived"].value_counts()

print("1) Cantidad de personas en primera clase:")
print("No sobrevivieron:", conteo_surv.get(0, 0))
print("Sobrevivieron:", conteo_surv.get(1, 0))

print("\n2) Cantidad de hombres y mujeres en ambos casos:")

# Cantidad de hombres y mujeres según sobrevivieron o no
tabla_sex = (
    primera
    .groupby(["Survived", "Sex"])
    .size()
    .unstack(fill_value=0)
)

print("\nNo sobrevivieron:")
print("Hombres:", tabla_sex.loc[0].get("male", 0))
print("Mujeres:", tabla_sex.loc[0].get("female", 0))

print("\nSobrevivieron:")
print("Hombres:", tabla_sex.loc[1].get("male", 0))
print("Mujeres:", tabla_sex.loc[1].get("female", 0))
```

**Pregunta 14:** Dado el dataset titanic:

* Se desea saber en que puerto abordaron más pasajeros de primera clase.

**Ejem:** Del total de 100 pasajeros en primera clase, en el "puerto 1" es donde abordaron más pasajeros de esta clase con un total de 60 pasejeros.

```python
# Leer dataset
df = pd.read_csv("titanic.csv")
df.columns = df.columns.str.strip()

# Filtrar pasajeros de primera clase
primera_clase = df[df["Pclass"] == 1]

# Contar pasajeros por puerto
conteo_puerto = primera_clase["Embarked"].value_counts()

# Puerto con más pasajeros
puerto_max = conteo_puerto.idxmax()
cantidad_max = conteo_puerto.max()

print("Pasajeros de primera clase por puerto:")
for puerto, cantidad in conteo_puerto.items():
    print(f"Puerto {puerto}: {cantidad}")

print(f"\nEl puerto donde abordaron más pasajeros de primera clase fue '{puerto_max}' con {cantidad_max} pasajeros.")
```
