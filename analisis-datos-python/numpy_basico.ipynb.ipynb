**Pregunta 1:** Definir una array en de 4x6 con elementos enteros multiplos de 4, a continuación imprimir:

* Numero de dimensiones
* Forma
* Tamaño total

```python
import numpy as np

# Crear array 4x6 con múltiplos de 4
array_4x6 = np.arange(4, 4*4*6 + 1, 4).reshape(4, 6)

# Imprimir resultados
print("Array:")
print(array_4x6)

print("Número de dimensiones:", array_4x6.ndim)
print("Forma:", array_4x6.shape)
print("Tamaño total:", array_4x6.size)
```

**Pregunta 3:** Dado el siguiente array en 3 dimensiones:
* Averiguar que forma tiene
* Acceder mendiante indices a los elementos remarcados en negrita. 

[[[  9  73  96  27  76]
  [ 90  **53**  44  72  72]
  [ 19  76  36  **37**   3]
  [ 78  89  84  21  69]]

 [[ 58   5  80  95  93]
  [ 39 **100**  24  81  28]
  [ 71  63  16  72  39]
  [ **83**  81  38  78 100]]

 [[ 84  **77**  35  35  26]
  [ 38   7  69  94  **29**]
  [ 40  85  82   9  48]
  [ 43  42 100 100  37]]]

```python
import numpy as np

arr = np.array([
    [
        [9, 73, 96, 27, 76],
        [90, 53, 44, 72, 72],
        [19, 76, 36, 37, 3],
        [78, 89, 84, 21, 69]
    ],
    [
        [58, 5, 80, 95, 93],
        [39, 100, 24, 81, 28],
        [71, 63, 16, 72, 39],
        [83, 81, 38, 78, 100]
    ],
    [
        [84, 77, 35, 35, 26],
        [38, 7, 69, 94, 29],
        [40, 85, 82, 9, 48],
        [43, 42, 100, 100, 37]
    ]
])

# Forma del array
print(arr.shape)

# Elementos remarcados en negrita
print(arr[0,1,1])  # 53
print(arr[0,2,3])  # 37
print(arr[1,1,1])  # 100
print(arr[1,3,0])  # 83
print(arr[2,0,1])  # 77
print(arr[2,1,4])  # 29
```

**Pregunta 5:** El último valor del array declarado no se imprime correctamente, evalue y solucione. A contuación explique a que se debe.   

```python
import numpy as np 

a = np.array([24,10,366],dtype=np.int8)
print(a)
```

```python
import numpy as np

a = np.array([24, 10, 366], dtype=np.int16)
print(a)
```

El último valor no se imprime correctamente porque el array se crea con el tipo np.int8,cuyo rango es de -128 a 127. El valor 366 excede ese rango y provoca overflow. Se soluciona utilizando un tipo de dato mayor como np.int16.

**Pregunta 7:** Utilizando nditer iterar sobre las filas el siguiente array. 

[ 58   5  80  95  93]

[ 39 100 24  81  28]

[ 71  63  16  72  39]

[ 83  81  38  78 100]

```python
import numpy as np

arr = np.array([
    [58,  5, 80, 95, 93],
    [39, 100, 24, 81, 28],
    [71, 63, 16, 72, 39],
    [83, 81, 38, 78, 100]
])

for fila in np.nditer(arr, flags=['external_loop','buffered'], 
                      buffersize=arr.shape[1], order='C'):
    print(fila)
```

**Pregunta 8:** Dada la siguiente matriz, convertirla en una matriz triangular inferior.

[58 92 88 84]

[ 5 23 90 31]

[97 32  7 94]

[87 84 10 23]

```python
import numpy as np

arr = np.array([
    [58, 92, 88, 84],
    [5,  23, 90, 31],
    [97, 32, 7,  94],
    [87, 84, 10, 23]
])

np.tril(arr)
```

**Pregunta 9:** Crear un array de tipo datetime con elementos entre 2010/10 y 2011/11.

**Pregunta 10:** Utilizando el array de la pregunta anterior se desea obtener la diferencia de cada una de las fechas con respecto al 07/10/2015. Obtener cada una de las diferencias y guardarlas en un array.

```python
import numpy as np

fechas = np.arange(
    np.datetime64('2010-10'),
    np.datetime64('2011-12'),
    dtype='datetime64[M]'
)

referencia = np.datetime64('2015-10-07')

diferencias = referencia - fechas
diferencias
```
