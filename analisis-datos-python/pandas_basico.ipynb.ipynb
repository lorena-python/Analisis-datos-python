**Pregunta 6:** Generar el data frame de la imagen bajo las siguientes reglas para los indices y columnas:

* Los indices deben ser creado mediante listas.
* Los indices de columnas mediante tuplas.

<center><img width="350"  src="df_6.png" /> <center/>

```python
import pandas as pd

# Datos
data = [
    [100, 300, 900, 400],
    [200, 500, 300, 600]
]

# Índices (LISTAS) -> valores r0/r1 y rA/rB
row_index = pd.MultiIndex.from_arrays(
    [
        ["r0", "r1"],   # Courses
        ["rA", "rB"]    # Fee
    ],
    names=["Courses", "Fee"]   # <- ESTO hace que se vea como en la imagen
)

# Columnas (TUPLAS)
columns = pd.MultiIndex.from_tuples([
    ("Gasoline", "Toyota"),
    ("Gasoline", "Ford"),
    ("Electric", "Tesla"),
    ("Electric", "Nio")
])

df6 = pd.DataFrame(data, index=row_index, columns=columns)
df6
```

**Pregunta 8:** Lea el dataset "Financial Sample.xlsx" y a partir de este crea un nuevo DF donde:

* La columna "Date" sea el indice de tipo datetime.
* Contenga las columnas Segment,Country,Product,Units Sold,Sales.


Para finalizar calcular la suma de la columna "Sales" e imprimirla.

```python
import pandas as pd

# Leer el dataset
df8 = pd.read_excel("Financial Sample.xlsx")

# Convertir la columna Date a datetime
df8["Date"] = pd.to_datetime(df8["Date"])

# Usar Date como índice
df8 = df8.set_index("Date")

# Seleccionar las columnas solicitadas 
df8 = df8[["Segment", "Country", "Product", "Units Sold", " Sales"]]

# Mostrar el Dataframe
df8

# Calcular e imprimir la suma de la columna Sales
df8[" Sales"].sum()
```

**Pregunta 9:** Ordena el DF de la pregunta anterior mediante el método sort_index a continuación calcula la diferencia entre el indice del primer elemento y el ultimo elemento.

```python
# Ordenar el DataFrame por el índice (Date)
df9 = df8.sort_index()

# Obtener la primera y la última fecha del índice
primera_fecha = df9.index[0]
ultima_fecha = df9.index[-1]

# Calcular la diferencia entre ambas fechas
diferencia = ultima_fecha - primera_fecha

diferencia
```
