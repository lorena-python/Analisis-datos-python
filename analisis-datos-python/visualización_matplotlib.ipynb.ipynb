**Pregunta 1:** Realiza la siguiente gráfica..

<center><img width="350"  src="pregunta1.png" /> <center/>

```python
# Colores del enunciado 
COLOR_VERDE = "#A9DFBF"
COLOR_AZUL  = "#AEB6BF"
COLOR_ROJO  = "#EC7063"

# Figura 
fig = plt.figure(
    figsize=(9,6),
    dpi=200,
    facecolor=COLOR_AZUL,     
    edgecolor=COLOR_ROJO,     
    linewidth=4
)

# Área del gráfico (verde)
ax = fig.add_subplot(111)
ax.set_facecolor(COLOR_VERDE)

# Configuración ejes 
ax.set_xlim(-10, 25)
ax.set_ylim(-150, 1200)

ax.set_xticks([-10, -5, 0, 5, 10, 15, 20, 25])
ax.set_yticks([0, 200, 400, 600, 800, 1000, 1200])

# Título 
plt.title("Figura")

# Bordes negros del área verde
for s in ax.spines.values():
    s.set_color("black")
    s.set_linewidth(1.2)

plt.show()
```

**Pregunta 2:** Graficar la función x^3-x^2+5 y configurarla como se muestra en la siguiente imagen.

Colores usados:
* #A9CCE3
* #A9DFBF
* #FAD7A0


<center><img width="350"  src="Screenshot_1.png" /> <center/>

```python
# Colores
COLOR_FIG   = "#A9CCE3"   # fondo exterior
COLOR_AX    = "#A9DFBF"   # fondo área gráfica
COLOR_LINEA = "#A9CCE3"   # línea función

# Datos 
x = np.linspace(-10, 10, 400)
y = x**3 - x**2 + 5

# Figura 
fig = plt.figure(
    figsize=(9,6),
    dpi=200,
    facecolor=COLOR_FIG,
    edgecolor="#FAD7A0",
    linewidth=3
)

ax = fig.add_subplot(111)
ax.set_facecolor(COLOR_AX)

# Gráfica 
plt.plot(x, y, color=COLOR_LINEA, linewidth=3)

# Configuración ejes 
ax.set_xlim(-10, 25)
ax.set_ylim(-100, 700)

ax.set_xticks([-10, -5, 0, 5, 10, 15, 20, 25])
ax.set_yticks([-100, 0, 100, 200, 300, 400, 500, 600, 700])

# Título y etiquetas 
plt.title("Gráfica pregunta 2")
plt.xlabel("Independiente")
plt.ylabel("Dependiente")

# Bordes negros del área
for s in ax.spines.values():
    s.set_color("black")
    s.set_linewidth(1.2)

plt.show()
```

**Pregunta 3:** Graficar las funciones:

* cos(x)
* 2cos(x)
* cos(2x)

Consideraciones:

* Centrar los ejes
* Limitar el ejeY  entre -3 y 3
* Limitar el ejeX entre -3pi y 3pi.
* Añadir los labels.  

```python
# Datos
x = np.linspace(-3*np.pi, 3*np.pi, 1000)

y1 = np.cos(x)
y2 = 2*np.cos(x)
y3 = np.cos(2*x)

# Figura
fig, ax = plt.subplots(figsize=(9, 5))

# Líneas
ax.plot(x, y1, label="cos(x)")
ax.plot(x, y2, label="2cos(x)")
ax.plot(x, y3, label="cos(2x)")

# Límites
ax.set_xlim(-3*np.pi, 3*np.pi)
ax.set_ylim(-3, 3)

# Centrar ejes 
ax.spines["left"].set_position("zero")
ax.spines["bottom"].set_position("zero")
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")

# Ticks y etiquetas
ax.set_xlabel("x")
ax.set_ylabel("y")

# Leyenda
ax.legend()

plt.show()
```

**Pregunta 4:** Utilizando el conjunto de datos "estudiantes.csv" mediante un  gráfico  de barras mostrar las tres notas de los primeros 7  alumnos con mayor promedio como se muestra acontinuación.

Configurar todo lo apreciado en la gráfica.

<center><img width="350"  src="Screenshot_2.png" /> <center/>

```python
# Leer archivo 
df = pd.read_csv(
    "estudiantes.csv",
    sep="-",
    header=0,
    names=["Indice", "Alumno", "Asignatura1", "Asignatura2", "Asignatura3"]
)

# Eliminar columna índice
df = df.drop(columns=["Indice"])

# Calcular promedio 
df["Promedio"] = df[["Asignatura1", "Asignatura2", "Asignatura3"]].mean(axis=1)

# Obtener top 7 alumnos 
top7 = df.sort_values(["Promedio", "Alumno"], ascending=[False, True]).head(7)

# Preparar datos 
alumnos = top7["Alumno"]
exam1 = top7["Asignatura1"]
exam2 = top7["Asignatura2"]
exam3 = top7["Asignatura3"]

x = np.arange(len(alumnos))
width = 0.25

fig, ax = plt.subplots(figsize=(6.5, 6), dpi=120)

ax.bar(x - width, exam1, width, label="Examen 1")
ax.bar(x,     exam2, width, label="Examen 2")
ax.bar(x + width, exam3, width, label="Examen 3")

ax.set_xticks(x)
ax.set_xticklabels(alumnos, fontsize=9)
ax.set_xlabel("Alumnos")
ax.set_ylabel("Notas")
ax.set_ylim(0, 120)

# Leyenda EXACTA como imagen 
ax.legend(loc="upper right", bbox_to_anchor=(1, 1.00))

# Ajuste manual de márgenes 
plt.subplots_adjust(left=0.10, right=0.95, top=0.90, bottom=0.12)

plt.show()



```

**Pregunta 6:** Utilizando el conjunto de datos "titanic.csv" mediante una gráfica circular mostrar la cantidad de pasajeros según el punto de embarque y si no existen datos también debe añadirlos a la gráfica.

```python
# Leer archivo 
df = pd.read_csv("titanic.csv") 

# Contar embarques (incluye NaN como categoría) 
conteo = df["Embarked"].fillna("Sin dato").value_counts()

# Asegurar que existan C, Q, S y Sin dato aunque no aparezcan 
orden = ["C", "Q", "S", "Sin dato"]
conteo = conteo.reindex(orden, fill_value=0)

# Gráfica circular
fig, ax = plt.subplots(figsize=(6, 6), dpi=120)

ax.pie(
    conteo.values,
    labels=conteo.index,
    autopct="%1.1f%%",
    startangle=90
)

ax.set_title("Pasajeros por punto de embarque")
plt.tight_layout()
plt.show()
```

**Pregunta 7:** Utilizando el conjunto de datos "Clientes.xlsx" y dentro de una sola figura configurar dos gráficas circulares:

* En la primera mostrar cuantas personas de genero másculino son solteras y casadas.
* En la segunda mostrar cuantas personas de genero femenino son solteras y casadas.

```python
df = pd.read_excel("Clientes.xlsx")

df.head()
```

```python
# Separar por género
masculinos = df[df["Género"] == "M"]
femeninos = df[df["Género"] == "F"]

# Contar estado civil
conteo_m = masculinos["EstadoCivil"].value_counts()
conteo_f = femeninos["EstadoCivil"].value_counts()

# Crear figura con 2 gráficos
fig, ax = plt.subplots(1, 2, figsize=(10,5))

# Gráfico masculinos
ax[0].pie(conteo_m.values,
          labels=conteo_m.index,
          autopct="%1.1f%%",
          startangle=90)
ax[0].set_title("Masculinos")

# Gráfico femeninos
ax[1].pie(conteo_f.values,
          labels=conteo_f.index,
          autopct="%1.1f%%",
          startangle=90)
ax[1].set_title("Femeninos")

plt.tight_layout()
plt.show()
```

**Pregunta 8:** Utilizando la primera tabla del conjunto de datos "Clientes.xlsx", en una figura  mostrar un gráfico por cada Nivel de "Eduación" donde se visualice mediante un gráfico de barras la cantidad de personas por ocupación.

Por ejemplo aqui tenemos los datos de nivel de educación bachiller:

<center><img width="350"  src="Screenshot_3.png" /> <center/>

```python
# Orden fijo de ocupaciones
orden_ocupaciones = [
    "Directivo",
    "Funcionario",
    "Militar",
    "Obrero Calificado",
    "Profesional"
]

# Niveles de educación
niveles = df["Educación"].unique()

fig, ax = plt.subplots(len(niveles), 1, figsize=(8, 4*len(niveles)))

if len(niveles) == 1:
    ax = [ax]

for i, nivel in enumerate(niveles):

    datos_nivel = df[df["Educación"] == nivel]
    conteo = datos_nivel["Ocupación"].value_counts()

    # Reordenar según el orden fijo
    conteo = conteo.reindex(orden_ocupaciones, fill_value=0)

    ax[i].bar(conteo.index, conteo.values)
    ax[i].set_title(f"Educación: {nivel}")
    ax[i].set_ylabel("Cantidad")
    ax[i].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()
    
```

**Pregunta 12:** Utilizando las dos tablas del conjunto de datos "Clientes.xlsx" se desea saber la distribución de personas según el pais.

* Las dos tablas esta unidas a través de la columna IdGeografia y CodGeografia.
* Puede ocupar bucles para unir las dos tablas.

```python
df_geo = pd.read_excel("Clientes.xlsx", sheet_name=1)
df_geo.head()
```

```python
df["Pais"] = ""
```

```python
for i in range(len(df)):
    
    codigo = df.loc[i, "CodGeografia"]
    
    for j in range(len(df_geo)):
        
        if codigo == df_geo.loc[j, "IdGeografia"]:
            
            df.loc[i, "Pais"] = df_geo.loc[j, "País"]
```

```python
conteo_pais = df["Pais"].value_counts()
display(conteo_pais)

plt.figure(figsize=(7,5))
conteo_pais.plot(kind="bar")
plt.title("Distribución de personas por país")
plt.xlabel("País")
plt.ylabel("Cantidad")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

**Pregunta 13:** Utilizando el dataset "titanic.csv" mostrar en un gráfico circular el conteo por clase.

Consideraciones:

* Resaltar la primera y tercera clase.
* Añadir sombra.

```python
# Cargar dataset
titanic = pd.read_csv("titanic.csv")

# Conteo por clase
conteo = titanic["Pclass"].value_counts().sort_index()

# Etiquetas
labels = ["Primera", "Segunda", "Tercera"]

# Resaltar primera y tercera
explode = [0.1, 0, 0.1]

# Gráfico circular
plt.figure(figsize=(7,7))

plt.pie(
    conteo.values,
    labels=labels,
    autopct="%1.1f%%",
    explode=explode,
    shadow=True,
    startangle=90
)

plt.title("Distribución de pasajeros por clase")
plt.axis("equal")   
plt.show()
```

**Pregunta 14:** Utilizar el resultado obtenido en la pregunta 11 del Examen4 y gráficarlo en un solo gráfico de barras horizontales donde se aprecie el conteo de mujeres y hombres por clase.

```python
titanic_df = pd.read_csv("titanic.csv")

df_sexo_clase = (
    titanic_df
    .groupby(["Pclass", "Sex"])
    .size()
    .unstack(fill_value=0)
)
```

```python
ax = df_sexo_clase.plot(kind="bar", figsize=(10,5), rot=0)

ax.set_title("Cantidad de hombres y mujeres por clase")
ax.set_xlabel("Clase")
ax.set_ylabel("Cantidad")
ax.legend(title="Sexo")

plt.tight_layout()
plt.show()
```
